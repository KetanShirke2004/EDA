```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
titanic=pd.read_csv('Titanic-Dataset.csv')
titanic.head()
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=tips)
```

```python
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)    
```

```python
sns.displot(titanic[titanic['Survived']==1]['Age'],kde=True)
sns.displot(titanic[titanic['Survived']==0]['Age'],kde=True)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived'] ))
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.lineplot(x='year',y='passengers',data=flights)
```

```python
sns.clustermap(flights.pivot_table(index='month',columns='year',values='passengers'),annot=True,fmt='d')
```
