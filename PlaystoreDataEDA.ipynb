```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
#covert eviw Datatype to int 
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')

```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year

```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
import os
os.makedirs('Clean_Dataset', exist_ok=True)
df_copy.to_csv('Clean_Dataset/google_clean_dataset.csv')
```

```python
df_copy[df.duplicated('App')].shape
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
numeric_feacture=[feature for feature in df_copy.columns if df_copy[feature].dtype !='object']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype =='object']
print('We have {} numerical features :{}'.format(len(numeric_feacture),numeric_feacture))
print('\nwe have {} categorical features: {}'.format(len(categorical_features),categorical_features))
```

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('---------------------------')
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
numeric_features_plot = [f for f in numeric_feacture if f != 'Last Updated']
for i in range(0,len(numeric_features_plot)):
    plt.subplot(4,2,i+1)
    sns.kdeplot(x=df_copy[numeric_features_plot[i]],shade=True,color='r')
    plt.xlabel(numeric_features_plot[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis of Categorical Features ',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
